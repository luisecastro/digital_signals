###Â Signal Generator

This function generates a signal using a trigonometric function:
- Amp: Amplitude of the signal. (Volts, Amperes, Watts, etc.)
- Phase: Phase of the signal at 0. (degrees)
- Frequency: Frequency of the signal or number of cycles per time unit. (hertz)
- Length: The number of complet periods of the signal to generate. (seconds)
- SR: Sampling rate, the number of samples per cycle we take of the signal. (# of samples/second)

```python
def s_gen(amp=1.,phase=0.,freq=1,func=np.sin,length=1,sr=44100): 
    t = np.linspace(1/sr,length,sr*length)                       
    return amp*func(2*np.pi*(freq*t+phase/360))      
```

###Â Fast Fourier Transform

This function performs the Fast Fourier Transform of the samples of the signal provided, it then normalizes using the maximum value to scale down the rest and returns all the values that are above a certain threshold.
- x: The signal to analyse
- threshold: The threshold value from 0.0 to 1.0
- sr: The sampling rate

```python
def s_fft(x,threshold=0.1,sr=44100):                                          
    result = pd.Series(abs(np.fft.fft(x)))
    length = len(result)
    result = result/result.max()
    result = result.ix[result>=threshold].ix[:int(len(result)/2)]
    result = result.sort_values(ascending=False)
    
    result.index /= length/sr
    
    return result                                               
```

When we analyze an exact number of cycles everything is clean and nice.

For the high_freq signal, 16000 cycles in 1 second and 44100 samples per second.

```python
s_fft(high_freq)
```

###Â Windowing

If we don't analyse full cycles, frequency leakeage happens.

```python
s_fft(high_freq[:42500])
```

Most references to the Hanning, Hamming and Blackman windows come from the signal processing literature, where it is used as one of many windowing functions for smoothing values. It is also known as an apodization (which means âremoving the footâ, i.e. smoothing discontinuities at the beginning and end of the sampled signal) or tapering function.

We directly multiply the sample of the signal by a the chosen window of the same length. Important to remember that a multiplication in time means a convolution in frequency.

```python
s_fft(high_freq[:42500]*np.blackman(42500))
```

```python
s_fft(high_freq[:42500]*np.hamming(42500))
```

```python
s_fft(high_freq[:42500]*np.hanning(42500))
```

We can see a greatly reduced frecuency leakage using the windows.

###Â Alias

According to the Nyquist Rate, our sample rate should be > than frequency of the highest frequency we want to recover.

For a 22 kHz signal, the minimum sampling rate to recover it is 44 kHz (which happens to be the CD's sampling rate, or CD quality).

If we have a signal:
- freq = 10 kHz
- Sampling = 25 kHz

```python
sr = 25e3
fr = 10e3
s_fft(s_gen(freq=fr,sr=sr),sr=sr)
```

But if we reduce the sampling rate or increase the frequency, alias occurs (and that's bad)

```python
sr = 12e3
s_fft(s_gen(freq=fr,sr=sr),sr=sr)
```

```python
sr = 25e3
fr = 42e3
s_fft(s_gen(freq=fr),sr=sr)
```

###Â Zero padding

Zero padding increases resolution by interpolation, when bins are not multiples of sampling frequency we might miss the exact frequencies.

The following function pads a signal with front and back zeros (half & half) and up to the closest superior power of 2.

FFT consist of different algorithms that are efficient implementations of DFT (Discrete Fourier Transform). 

They have a O(nLogn) instead of O(n^2) and work better with power of 2 samples.

```python
def padder(x,power=0):                                                
    pad   = int(2**np.ceil(np.log2(len(x))+power))           
    zeros = pad-len(x)                                       
    x   = np.pad(x,mode='constant',pad_width=int(zeros/2))   
    return x                                                 
```

**Plot**
- FFT
- FFT + Zero Padding (increasing resolution)
- FFT + Hamming Window (reducing frequency leakage)
- FFT + Hamming Window + Zero Padding

```python
x = s_gen(freq=6,length=3.25)

plt.plot(s_fft(x,threshold=0).sort_index().ix[:10],alpha=1)
plt.plot(s_fft(padder(x,power=3),threshold=0).sort_index().ix[:10],alpha=0.5)
plt.plot(s_fft(np.hamming(143325)*x,threshold=0).sort_index().ix[:10],alpha=0.5)
plt.plot(s_fft(padder(np.hamming(int(44100*3.25))*x,power=3),threshold=0).sort_index().ix[:10],alpha=1)
plt.legend(['FFT','FFT+ZP','FFT+HW','FFT+HW+ZP'])
plt.show()
```

###Â Spectrograms

Spectrograms are 2D heatmap plots, with the x axis being the time dimension and the y axis being the frequency dimension. 

- **Fs or Sampling frequency:** should be specified and should be the same used to record the the audio.
Human hearing range is from around 20 Hz to 20k Hz. For cd audio quality a sampling rate of 44100 is used (twice the upper limit of the human range) this is due to Nyquist rate, that tells that sampling should at least twice the highest frequency you want to recover. If the sampling rate is inferior to the requested Nyquist frequency, the components that cannot be represented are folded, causing an erroneous spectral representation : this phenomenon is called "aliasing" or spectral folding.

- **NFFT:** Analysis of the signal for the spectrogram is done with a series of (typically overlaping) windows. By specifying the length of the windows we inccur in a tradeoff. Big windows will be able to include more frequencies (particularly lower ones) as it takes a bigger chunk of the signal to analyse (and allows for low frequencies to be included) but that gives us uncertainty on the exact position in time of those frequencies. The window specifies the number of samples to use. The lower the frequency, the larger the window needs to be (1/freq). WS = (Fs/freq) [1](http://support.ircam.fr/docs/AudioSculpt/3.0/co/Window%20Size.html)

- **STFT:** The method to transform a signal from the time domain to the frequency domain for the spectrogram is called the Fourier transform, in this case since we are interested in the frequency content of a signal along time, chunks of the signal are analysed one by one.  Thus the Short Time Fourier transform which is a Fast Fourier Transform for each chunk.

- **FFT:** Fast Fourier Transform is and optimized version of the DFT (Discrete Fourier Transform), whereas the DFT takes O(n^2) time, the FFT takes only O(nlog(n)). There are multiple versions of the FFT.

Knowledge of the type of frequencies expected can help to select those parameters. [2](https://stackoverflow.com/questions/29321696/what-is-a-spectrogram-and-how-do-i-set-its-parameters)

```python
import librosa
```

```python
rotv, sr = librosa.load('rotv.m4a',sr=44100)
IPython.display.Audio(rotv,rate=sr)
```

```python
_ = plt.specgram(rotv,Fs=sr,NFFT=2048,cmap=plt.get_cmap('Spectral'))
plt.colorbar(format='%+2.0f dB')
plt.ylim(0,22100)
plt.title("Spectrogram")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (seconds)")
plt.show()
```

```python
_ = plt.specgram(rotv,Fs=sr,NFFT=256,cmap=plt.get_cmap('Spectral'),)
plt.colorbar(format='%+2.0f dB')
plt.ylim(0,22100)
plt.title("Spectrogram")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (seconds)")
plt.show()
```

Compare both spectrograms, the first one has a 2048 samples window, the second one 256. The first with the bigger window offers better frequency definition, identifying more clearly the exact frequencies while not being as accurate in the time department as the 256 window.

The size of the window is also relevant to identify lower frequencies, the bigger the better lower frequencies (a bigger portion or an entire cycle of the signal can fit in the window.)

**Naive classifier**

Split data in training and test sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(voice['meanfun'],target,train_size=0.6,random_state=0)
```

Create the class naive classifier which will take the training set which is just one feature, the highest correlated with the target and assuming a Gaussian distribution for each class will assign a class on the test set.

```python
class naive_class(object):
    def __init__(self):
        pass
    
    def fit(self,X_train,y_train):
        self.unique = y_train.unique()
        self.mm = X_train.ix[target==self.unique[0]].mean()
        self.ms = X_train.ix[target==self.unique[0]].std()
        self.fm = X_train.ix[target==self.unique[1]].mean()
        self.fs = X_train.ix[target==self.unique[1]].std()
      
    def predict(self,X_test):         
        mc = abs((X_test-self.mm)/self.ms)
        fc = abs((X_test-self.fm)/self.fs)
        
        result = []
        for i,j in zip(mc,fc):     
            if i <= j:
                result.append(self.unique[0])
            else:
                result.append(self.unique[1])
                
        return result
```

```python
clf = naive_class()
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix as CF

def cm_stats(cf):
    acc = 0.
    for i in range(cf.shape[0]):
        acc += cf[i][i]
    print('Accuracy: {0:.4f}'.format(acc/sum(sum(cf))))
    print('Precision: {0:.4f}'.format(cf[0][0]/(cf[0][0]+cf[1][0])))
    print('Recall: {0:.4f}'.format(cf[0][0]/(cf[0][0]+cf[0][1])))
```

```python
cm_stats(CF(y_test,clf.predict(X_test)))
```

###Â Filters

**Butterworth filter**

Digital filters (and their analog counterparts) are designed to limit certain frequencies and allow to pass others.
- Pass band: Allows a certain range of frequencies
- Stop band: Rejects a certain range of frequencies
- Low pass: Allows for frequencies under a certain threshold
- High pass: Allows for frequencies over a certain threshold

Typically the filters are designed to have a -3 dB or 1/sqrt(2) response on the threshold frequencies. They can be differentiated also on how flat are their pass and reject bands and how steep are the transitions from one to the other.

The Butterworth filter takes and additive approach (degrees) and is considered for the flatness of its pass band.

```python
from scipy.signal import butter, lfilter
```

The cutoff frequency for the digital filter is specified as a number in the range [0,1] where 1 is the Nyquist frequency.

```python
degree = 4
cutoff = 0.03125 # 44100 * 0.03125 = 1378
ts    = pd.Series(x[:4410]+s_gen(freq=1400,amp=2,length=0.1))
```

```python
ts.plot(alpha=0.5)
pd.Series(data = lfilter(*butter(degree,cutoff, 'low'), x = ts.values),index = ts.index).plot(alpha = 1.0).plot()
plt.legend(['data','Butterworth '+str(degree)+'th order IIR '+str(cutoff)])
plt.show()
```

IIR is for an Infinite Impulse Response filter, FIR is Finite Impulse Response their main difference is that the former creates the output signal using the previous input signals and outputs whereas the latter just uses previous outputs. Other concerns are stability as an IIR has a feedback loop.

```python
s_fft(ts)
```

```python
s_fft(lfilter(*butter(degree,cutoff,'low'),x = ts.values))
```

**Power Spectral Density**

The power spectral density  PxxPxx  of a signal is an estimate of the power in each frequency component. It can be obtained by averaging the FFT obtained in several frames (Welch's method). The signal is divided into NFFT length segments. Each segment is detrended by function detrend and windowed by function  window. noverlap gives the length of the overlap between segments. The  |fft(i)|2|fft(i)|2  of each segment  ii  are averaged to compute  PxxPxx , with a scaling to correct for power loss due to windowing.

```python
_, _ = plt.psd(rotv, Fs = 44100)
_, _ = plt.psd(lfilter(*butter(degree,cutoff*2,'low'),x = rotv), Fs = 44100)
plt.show()
```
